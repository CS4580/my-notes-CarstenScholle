### Task: Understand Museum Attendance


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the data
data_path = '../data/museum_visitors.csv'
museum_df = pd.read_csv(data_path, index_col='Date', parse_dates=True)

museum_df.head()
```

The Firehouse Museum claims that thy ran an event in 2014 that brought a big number of visitors, and that they should get extra budget to run a similar event this year.

Task: Show the museum board how the event compared to regular traffic at each museum.
- Create a `lineplot chart` that shows how the number of visitors to change on each museum over time 

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=museum_df)
plt.title("Daily number of visitors in museums over time")
plt.ylabel("Number of visitors")
```

### Assess Seasonality
When meeting with the employees of the `Avila Adobe`, you hear that ine major pain point is that the number of visitors varies greatly with the season.

Part A:
- Create a `Line chart` that shows the number of visitors to the `Avila Adobe` has evolved over time
- Replace `inf` with `nan`, and then fill them in with `0`

```python
import numpy as np
# Replace infinite values with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)
# Now we can use the fillna() method to replace NaN with zeros
museum_df = museum_df.fillna(0)
```

```python
# Plot the data
plt.figure(figsize=(16, 6))
sns.lineplot(data=museum_df['Avila Adobe'], label='Avila Adobe')
plt.title('Monthly Visitors to Avila Adobe')
plt.xlabel('Date')

```

## Bar Plots

Dataset: US Department of Transportation that tracks flight delays

```python
# load dataset
file_path = '../data/flight_delays.csv'
flight_delay_df = pd.read_csv(file_path, index_col='Month')
flight_delay_df.head()
```

```python
# Use a bar plot
plt.figure(figsize=(16, 6))
plt.title('Average Arrival Delay for Spirit Airlines, by Month')
sns.barplot(x=flight_delay_df.index, y=flight_delay_df['NK'])
plt.ylabel('Arrival Delay in Minutes')
```

# Heat Maps

```python
# Plot heatmap for all data
plt.figure(figsize=(16, 6))
plt.title('Avg Arrival Delay for Each Airline, by Month')
sns.heatmap(data=flight_delay_df, annot=True)
plt.xlabel('Airline')
```

## Scatter Plots

### Load the Dataset
We will use a (synthetic) dataset of medical insurance charges: `data/insurance.csv`

```python
df_insurance = pd.read_csv('../data/insurance.csv')
df_insurance.head()
```

Create a `scatter` plot. In seaborn use `sns.scatterplot` command.
- x-axis use `bmi`
- y-axis use `charges`

```python
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'])
```

The plot suggest we have a `positive correlation` between BMI and charges.
- Higher BMI tends to pay more

To double check the strength of this relationship, add a `regression line` that best fits the data

```python

sns.regplot(x=df_insurance['bmi'], y=df_insurance['charges'])
```

### Color-coded Scatter Plots
They are good when you need to display the correlation between not two, but more variables.


We want to understand how `smokers` are affects the relationship between `bmi` and `charges`

```python
# Scatter plot of bmi vs charges AND smoker vs charges
# x = bmi, y = charges, hue = smoker
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'], hue=df_insurance['smoker'])
```

```python
# Add two regression lines
# # When using lmplot for three or more, only use the column name 
sns.lmplot(data=df_insurance, x='bmi', y='charges', hue='smoker')
```

Finally, we can adapt the design of a scatter plot to feature a `categorical variable` (like 'smoker') on one of the main axis. Some people call this plot: **Categorical Scatter Plot**

```python
sns.swarmplot(x=df_insurance['smoker'], y=df_insurance['charges'])
```
